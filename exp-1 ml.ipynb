```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
df =pd.read_csv("CarPrice_Assignment.csv")
df.head()
```

```python
x=df[['enginesize','horsepower','citympg','highwaympg']]
y=df['price']
#select features and target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#split data
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
#feature scaling
```

```python
model=LinearRegression()
model.fit(x_train_scaled,y_train)
#training the model
```

```python
#prediction
y_pred=model.predict(x_test_scaled)
```

```python
print('Name: KIRAN MUKESH ')
print('Reg. No: 21225040188 ')
print('MODEL COEFFICIENTS')
for feature,coef in zip(x.columns,model.coef_):
    print(f"{feature:}: {coef:}")
print(f"{'Intercept':}: {model.intercept_:}")
```

```python
print("\nMODEL PERFORMANCE: ")
print(f"{'MSE':>12}: {mean_squared_error(y_test,y_pred):>10.2f}")
print(f"{'RMS':>12}: {np.sqrt(mean_squared_error(y_test,y_pred)):}")
print(f"{'R-Squared':>12}: {r2_score(y_test,y_pred):}")
print(f"{'MAE':>12}: {(mean_absolute_error(y_test,y_pred)):}")

      
```

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test,y_pred,alpha=0.6)
plt.plot([y.min(),y.max()],[y.min(),y.max()],'r--')
plt.title("Linerity Check: Actual vs Predicted Prices")
plt.xlabel("Actual Price ($)")
plt.ylabel("Predicted Price ($)")
plt.grid(True)
plt.show()
#Linearity check
```

```python
#Independance (Durbin-Watson)
residuals=y_test-y_pred
dw_test=sm.stats.durbin_watson(residuals)
print(f"\nDurbin-Watson Statistic: {dw_test:.2f}",
      "\n(Values close to 2 indicate no autocorrelation)")
```

```python
#Homoscdasticty
plt.figure(figsize=(10,5))
```

```python
#Homoscedasticity
plt.figure(figsize=(10,5))
sns.residplot(x=y_pred,y=residuals,lowess=True,line_kws={'color' : 'red'})
plt.title("Homoscedasticity Check: Residuals vs Predicted")
plt.xlabel("Predicted Price ($)")
plt.grid(True)
plt.show()
```

```python
#Normality of residuals
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,5))
sns.histplot(residuals,kde=True,ax=ax1)
ax1.set_title("Residuals Distribution")
sm.qqplot(residuals,line='45',fit=True,ax=ax2)
ax2.set_title=("Q-Q Plot")
plt.tight_layout()
plt.show()
```
